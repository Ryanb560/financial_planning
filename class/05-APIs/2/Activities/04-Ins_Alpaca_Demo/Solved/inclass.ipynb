```python
import os 
import pandas as pd
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi
```

```python
load_dotenv()
```

```python
alpaca_key = os.getenv("ALPACA_API_KEY")
alpaca_secret = os.getenv("ALPACA_SECRET_KEY")
```

```python
alpaca = tradeapi.REST(
    alpaca_key,
    alpaca_secret,
    api_version="v2"
)
```

```python
alpaca.get_barset(
    ["GME","AMC","GE"],
    timeframe="1D"
    ).df
```

```python
today = pd.Timestamp("2021-04-08", tz="America/Denver").isoformat()
today
```

```python
df_portfolio = alpaca.get_barset(
    ["FB","AAPL","XOM","ENPH","GOOG", "MSFT","MU","JNJ","SPWR"],
    timeframe="1D",
    end=today,
    limit=1000
).df

df_portfolio.head()
```

```python
df_portfolio["AAPL"]["close"]
```

```python
df_portfolio["GOOG"]["close"]
```

```python
df_closing_prices = pd.DataFrame()

df_closing_prices["AAPL"] = df_portfolio["AAPL"]["close"]
df_closing_prices["GOOG"] = df_portfolio["GOOG"]["close"]
df_closing_prices.index = df_closing_prices.index.date
df_closing_prices.head()
```

```python
closing_price_fast=df_portfolio.xs("close", axis="columns", level=1)
closing_price_fast.index = closing_price_fast.index.date
closing_price_fast.head()
```

```python
returns = closing_price_fast.pct_change().dropna()
returns.head()
```

```python
returns.plot()
```
