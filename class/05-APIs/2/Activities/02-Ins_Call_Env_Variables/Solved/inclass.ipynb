```python
from dotenv import load_dotenv
import os
import requests
```

```python
load_dotenv()
```

```python
api_key = os.getenv("QUANDL_API_KEY")
print(type(api_key))
```

```python
url = "https://www.quandl.com/api/v3/datasets/WIKI/AAPL.json"

querystring = f"?api_key={api_key}"
```

```python
url+querystring
```

```python
res = requests.get(url+querystring).json()
```
