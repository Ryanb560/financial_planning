```python
from dotenv import load_dotenv
import os
import requests
from pprint import pprint
load_dotenv()
```

```python
api_key = os.getenv("QUANDL_API_KEY")
print(type(api_key))
```

```python

url = "https://www.quandl.com/api/v3/datasets/WIKI/AAPL.json"
params_dict = {
    "api_key": api_key
}

res = requests.get(url, params=params_dict).json()
pprint(res)
```

```python
def get_stock_data(ticker):
    url = f"https://www.quandl.com/api/v3/datasets/WIKI/{ticker}.json"
    params_dict = {
        "api_key": api_key
    }

    res = requests.get(url, params=params_dict)
    if res.status_code == 404:
        return {
            "message": "That stock doesn't exist!!"
        }

    if res.status_code == 200:
        return res.json()
```

```python
get_stock_data("GOOGL")
```
