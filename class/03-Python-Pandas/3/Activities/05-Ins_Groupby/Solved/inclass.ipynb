```python
import pandas as pd
from pathlib import Path
```

```python
csv_path = Path('../Resources/crypto_data.csv')
crypto_data = pd.read_csv(
    csv_path, 
    index_col='data_date', 
    parse_dates=True, 
    infer_datetime_format=True)


crypto_data
```

```python
crypto_data = crypto_data.drop(columns=["timestamp","data_time"])
crypto_data
```

```python
crypto_data = crypto_data.sort_index()
crypto_data
```

```python
crypto_data.dropna(inplace=True)
crypto_data.head()
```
